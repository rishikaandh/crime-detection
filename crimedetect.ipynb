```python
import time
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderUnavailable

geolocator = Nominatim(user_agent="geoapiExercises")

def get_country_from_coordinates(row):
    lat, lon = row['Latitude'], row['Longitude']
    location = None
    retries = 0

    while retries < 3:  # Retry up to 3 times
        try:
            location = geolocator.reverse((lat, lon), exactly_one=True)
            break  # Break out of the loop on success
        except GeocoderTimedOut as e:
            print(f"Geocoding timed out. Retrying ({retries + 1}/3)...")
            retries += 1
            time.sleep(1)  # Sleep for a second before retrying
        except GeocoderUnavailable as e:
            print("Geocoding service is currently unavailable. Retrying in a moment...")
            time.sleep(10)  # Sleep for 10 seconds before retrying

    if location:
        country = location.raw.get('address', {}).get('country', 'Unknown')
        return country
    else:
        return 'Unknown'
```

```python
# Calculate the total number of rows and the part size
total_rows = len(main)
part_size = total_rows // 10000  # Divide the data into 100 equal parts

# Initialize a list to store the parts
data_parts = []

# Split the data into 100 equal parts and store them in the list
for i in range(10000):
    start_idx = i * part_size
    end_idx = (i + 1) * part_size if i < 99 else total_rows
    part = main.iloc[start_idx:end_idx]
    data_parts.append(part)
```
